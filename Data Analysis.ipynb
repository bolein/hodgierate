```python
import data.loader as loader
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('data/final_set.csv')
data['encode'] = data['encode'].apply(lambda row: np.asarray([int(c) for c in row[1:-1].split(', ')]))
print data.Class[data.Class == 1].count() + data.Class[data.Class == 0].count()
```

```python
good = np.hstack(data.encode[data.Class == 0].values)
bad = np.hstack(data.encode[data.Class == 1].values)
```

```python
plt.figure(figsize=(15, 10))
plt.hist(good, bins=128, label='good', alpha=0.5)
plt.hist(bad, bins=128, label='bad', alpha=0.5)
plt.legend(prop={'size':25})
plt.xlim(0, 128)
plt.show()
```
